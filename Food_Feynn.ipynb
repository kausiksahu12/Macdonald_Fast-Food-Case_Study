EXTRACTING SEGMENTS

```python
from sklearn.cluster import KMeans
# Select the first 11 columns and convert to matrix
# Convert "Yes" to 1 and "No" to 0
# Set random seed
np.random.seed(1234)

# Perform KMeans clustering
results = {}
for n_clusters in range(2, 9):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10)
    kmeans.fit(mn)
    results[n_clusters] = kmeans.labels_

# Relabel the clusters
for n_clusters, labels in results.items():
    unique_labels = pd.unique(labels)
    relabeled_labels = range(n_clusters)
    for i, label in enumerate(unique_labels):
        labels[labels == label] = relabeled_labels[i]

# Print the relabeled clusters
for n_clusters, labels in results.items():
    print(f"Clusters with {n_clusters} centers:")
    print(labels)
```

```python
num_segments = []
labels = []
for n_clusters, cluster_labels in results.items():
    num_segments.append(n_clusters)
    labels.append(cluster_labels)
plt.figure(figsize=(10, 6))
plt.plot(num_segments, [len(set(l)) for l in labels], marker='o', linestyle='-')
plt.xlabel('Number of Segments')
plt.ylabel('Number of Clusters')
plt.title('Number of Clusters vs. Number of Segments')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import numpy as np
import matplotlib.pyplot as plt
# set the random seed
np.random.seed(1234)

# perform K-means clustering with 2 to 8 clusters and 10 repetitions
n_clusters = range(2, 9)
n_reps = 10
scores = []
for k in n_clusters:
    km = KMeans(n_clusters=k, max_iter=300, n_init=n_reps).fit(mn)
    km_labels = km.labels_
    score_k = []
    for i in range(n_reps):
        km_rep = KMeans(n_clusters=k, max_iter=300, n_init=1).fit(mn)
        km_rep_labels = km_rep.labels_
        score_k.append(adjusted_rand_score(km_labels, km_rep_labels))
    scores.append(score_k)

# plot the adjusted Rand index scores for each number of clusters
plt.errorbar(n_clusters, np.mean(scores, axis=1), yerr=np.std(scores, axis=1), fmt="o-")
plt.xlabel("number of clusters")
plt.ylabel("adjusted Rand index")
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.utils import resample

# Set random seed
np.random.seed(1234)

# Dictionary to store bootstrapped clustering results
boot_clusters = {}

# Iterate over different numbers of segments
for n_clusters in range(2, 9):
    # List to store bootstrapped labels
    boot_labels = []

    # Perform bootstrapping
    for _ in range(100):
        # Resample data with replacement
        boot_data = resample(mn, replace=True, n_samples=len(mn))

        # Perform KMeans clustering
        kmeans = KMeans(n_clusters=n_clusters, n_init=10)
        kmeans.fit(boot_data)

        # Store labels
        boot_labels.append(kmeans.labels_)

    # Store bootstrapped clustering results
    boot_clusters[n_clusters] = boot_labels

# Plotting adjusted Rand index against the number of segments
num_segments = list(boot_clusters.keys())
adjusted_rand_indices = []

for n_clusters, boot_labels in boot_clusters.items():
    # Compute adjusted Rand index for each bootstrapped clustering
    rand_indices = []
    for i in range(100):
        for j in range(i + 1, 100):
            rand_index = adjusted_rand_score(boot_labels[i], boot_labels[j])
            rand_indices.append(rand_index)
    # Average Rand index over all pairs of bootstrapped clusterings
    avg_rand_index = np.mean(rand_indices)
    adjusted_rand_indices.append(avg_rand_index)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(num_segments, adjusted_rand_indices, marker='o', linestyle='-')
plt.xlabel('Number of Segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Adjusted Rand Index vs. Number of Segments')
plt.grid(True)
plt.show()

```

```python
import numpy as np
from sklearn.mixture import GaussianMixture

# Set random seed
np.random.seed(1234)

# Create Gaussian Mixture Model object
n_components_range = range(2, 9)
gmm_models = []

# Fit GMM for different numbers of components
for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, covariance_type='full', n_init=10)
    gmm.fit(mn)
    gmm_models.append(gmm)

# Print model results
for n_components, gmm in zip(n_components_range, gmm_models):
    print(f"Number of components: {n_components}")
    print(f"Converged: {gmm.converged_}")
    print(f"Log Likelihood: {gmm.lower_bound_}")
    print(f"AIC: {gmm.aic(mn)}")
    print(f"BIC: {gmm.bic(mn)}")
    print("------------------------------")

```

```python
import matplotlib.pyplot as plt

# Extract AIC, BIC, and ICL values from the model results
AIC_values = [gmm.aic(mn) for gmm in gmm_models]
BIC_values = [gmm.bic(mn) for gmm in gmm_models]
ICL_values = [-gmm.lower_bound_ for gmm in gmm_models]  # Note: In FlexMix, ICL is negative log likelihood

# Plotting
plt.plot(n_components_range, AIC_values, label='AIC', marker='o')
plt.plot(n_components_range, BIC_values, label='BIC', marker='o')
plt.plot(n_components_range, ICL_values, label='ICL', marker='o')

plt.xlabel('Number of Components')
plt.ylabel('Value of Information Criteria')
plt.title('Value of Information Criteria vs. Number of Components')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
import numpy as np

# Assuming MD_x contains your data

# Calculate the distance matrix
distance_mat = distance_matrix(np.transpose(mn), np.transpose(mn))

# Perform hierarchical clustering
MD_vclust = hierarchy.linkage(distance_mat, method='complete')  # You can choose different methods here

print(MD_vclust)
```

```python
import matplotlib.pyplot as plt

# Assuming MD_k4 contains the clustering results
# Assuming MD_vclust contains the hierarchical clustering results

# Define the order of the clusters based on the hierarchical clustering
order = MD_vclust[:, 3].astype(int) - 1  # MD_vclust[:, 3] contains the order

# Get the counts of each cluster
cluster_counts = [sum(kmeans.labels_ == i) for i in range(max(kmeans.labels_) + 1)]

# Ensure that the order and cluster_counts variables have the same length
if len(order) != len(cluster_counts):
    if len(order) < len(cluster_counts):
        order = np.append(order, [0] * (len(cluster_counts) - len(order)))
    elif len(order) > len(cluster_counts):
        order = order[:len(cluster_counts)]

# Create the barchart with shading based on the order
plt.bar(range(len(cluster_counts)), cluster_counts, color='skyblue')
plt.bar(order, cluster_counts, color='lightgrey')  # Add shading based on the order

plt.xlabel('Cluster')
plt.ylabel('Count')
plt.title('Barchart of Clusters with Shading based on Hierarchical Clustering Order')

plt.show()
```
